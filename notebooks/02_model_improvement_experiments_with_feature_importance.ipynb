# 02 — Model Improvement Experiments + Feature Importance

Notebook ini tetap memakai pipeline eksperimen model sebelumnya: multiclass risk level, binary at-risk, oversampling, holdout report, dan leakage demo.

Revisi tambahan pada notebook ini adalah bagian **feature importance** untuk menjawab pertanyaan: fitur apa yang benar-benar memengaruhi prediksi model valid, dan kenapa skor bisa menjadi sangat tinggi ketika tiga fitur penentu (`depression`, `anxiety`, `panic_attack`) dimasukkan sebagai fitur.

Tujuan utamanya bukan memaksa skor tinggi, tetapi mencari pendekatan modeling yang paling masuk akal untuk dataset kecil serta menjelaskan batasan model secara jujur.

## 1. Import Library

```python
import warnings
warnings.filterwarnings("ignore")

from pathlib import Path
import numpy as np
import pandas as pd

from sklearn.model_selection import StratifiedKFold, cross_validate, train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.metrics import classification_report, accuracy_score, f1_score, balanced_accuracy_score

from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, GradientBoostingClassifier

from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import RandomOverSampler, SMOTE
```

## 2. Load dan Cleaning Dataset

Cleaning dilakukan supaya format data lebih konsisten sebelum masuk ke proses modeling.

```python
DATA_PATH = Path("../data/Student Mental health.csv")
if not DATA_PATH.exists():
    DATA_PATH = Path("data/Student Mental health.csv")

raw_df = pd.read_csv(DATA_PATH)

df = raw_df.rename(columns={
    "Timestamp": "timestamp",
    "Choose your gender": "gender",
    "Age": "age",
    "What is your course?": "course",
    "Your current year of Study": "year_of_study",
    "What is your CGPA?": "cgpa",
    "Marital status": "marital_status",
    "Do you have Depression?": "depression",
    "Do you have Anxiety?": "anxiety",
    "Do you have Panic attack?": "panic_attack",
    "Did you seek any specialist for a treatment?": "seek_treatment",
})

df["gender"] = df["gender"].fillna("Unknown").astype(str).str.strip().str.title()
df["course"] = df["course"].fillna("Unknown").astype(str).str.strip().str.replace(r"\s+", " ", regex=True).str.title()
df["age"] = pd.to_numeric(df["age"], errors="coerce")
df["age"] = df["age"].fillna(df["age"].median())
df["year_of_study"] = (
    df["year_of_study"].astype(str).str.extract(r"(\d+)")[0]
    .fillna("0").astype(int).map(lambda x: f"Year {x}" if x else "Unknown")
)
df["cgpa"] = df["cgpa"].fillna("Unknown").astype(str).str.strip().str.replace(r"\s+", " ", regex=True)

for col in ["depression", "anxiety", "panic_attack", "seek_treatment"]:
    df[col] = np.where(df[col].astype(str).str.strip().str.title() == "Yes", "Yes", "No")

df.head()
```

## 3. Membuat Target Multiclass dan Binary

Target multiclass tetap menggunakan `Low Risk`, `Medium Risk`, dan `High Risk`. Target binary menggabungkan `Medium Risk` dan `High Risk` menjadi `At Risk`, sehingga lebih cocok untuk screening awal.

```python
df["symptom_count"] = (
    (df["depression"] == "Yes").astype(int)
    + (df["anxiety"] == "Yes").astype(int)
    + (df["panic_attack"] == "Yes").astype(int)
)

df["risk_level"] = df["symptom_count"].map(
    lambda x: "Low Risk" if x == 0 else ("Medium Risk" if x == 1 else "High Risk")
)
df["at_risk"] = np.where(df["symptom_count"] > 0, "At Risk", "No Risk")

print("Jumlah data:", df.shape[0])
print("\nDistribusi risk_level:")
print(df["risk_level"].value_counts())
print("\nDistribusi target binary:")
print(df["at_risk"].value_counts())
```

## 4. Preprocessing, Fitur, dan Model

Fitur `depression`, `anxiety`, dan `panic_attack` tidak dipakai dalam model valid karena label risiko dibuat dari tiga indikator tersebut. Kalau dipakai, hasilnya akan tinggi tetapi bocor.

```python
FEATURES = ["age", "gender", "course", "year_of_study", "cgpa", "marital_status"]
NUM_FEATURES = ["age"]
CAT_FEATURES = ["gender", "course", "year_of_study", "cgpa", "marital_status"]

X = df[FEATURES]
y_multi = df["risk_level"]
y_binary = df["at_risk"]

preprocessor = ColumnTransformer([
    ("num", Pipeline([
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler())
    ]), NUM_FEATURES),
    ("cat", Pipeline([
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
    ]), CAT_FEATURES),
])

models = {
    "Dummy Most Frequent": DummyClassifier(strategy="most_frequent"),
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Logistic Regression Balanced": LogisticRegression(max_iter=1000, class_weight="balanced", random_state=42),
    "Decision Tree Balanced": DecisionTreeClassifier(max_depth=4, min_samples_leaf=3, class_weight="balanced", random_state=42),
    "Random Forest Balanced": RandomForestClassifier(n_estimators=40, max_depth=4, min_samples_leaf=3, class_weight="balanced", random_state=42),
    "Extra Trees Balanced": ExtraTreesClassifier(n_estimators=40, max_depth=4, min_samples_leaf=3, class_weight="balanced", random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scoring = {
    "accuracy": "accuracy",
    "balanced_accuracy": "balanced_accuracy",
    "f1_macro": "f1_macro",
    "f1_weighted": "f1_weighted",
}
```

## 5. Fungsi Evaluasi Cross-Validation

```python
def evaluate_models(X, y, experiment_name, model_dict):
    rows = []
    for model_name, model in model_dict.items():
        pipe = Pipeline([
            ("preprocess", preprocessor),
            ("model", model)
        ])
        scores = cross_validate(pipe, X, y, scoring=scoring, cv=cv, n_jobs=1)
        rows.append({
            "experiment": experiment_name,
            "model": model_name,
            "accuracy_mean": scores["test_accuracy"].mean(),
            "balanced_accuracy_mean": scores["test_balanced_accuracy"].mean(),
            "f1_macro_mean": scores["test_f1_macro"].mean(),
            "f1_weighted_mean": scores["test_f1_weighted"].mean(),
        })
    return pd.DataFrame(rows).sort_values("f1_macro_mean", ascending=False)
```

## 6. Eksperimen 1 — Multiclass Risk Level

```python
result_multi = evaluate_models(X, y_multi, "Multiclass Risk Level", models)
result_multi.round(3)
```

## 7. Eksperimen 2 — Binary At Risk

Hasil binary biasanya lebih stabil karena tugas model lebih sederhana: membedakan mahasiswa yang tidak punya indikator dan yang memiliki minimal satu indikator.

```python
result_binary = evaluate_models(X, y_binary, "Binary At Risk", models)
result_binary.round(3)
```

## 8. Eksperimen 3 — Oversampling untuk Binary Target

Oversampling dicoba hanya untuk target binary agar eksperimen tetap ringan dan relevan dengan sistem early warning.

```python
oversampling_models = {
    "LogReg Balanced + ROS": ImbPipeline([
        ("preprocess", preprocessor),
        ("oversample", RandomOverSampler(random_state=42)),
        ("model", LogisticRegression(max_iter=1000, class_weight="balanced", random_state=42)),
    ]),
    "LogReg Balanced + SMOTE": ImbPipeline([
        ("preprocess", preprocessor),
        ("oversample", SMOTE(random_state=42, k_neighbors=3)),
        ("model", LogisticRegression(max_iter=1000, class_weight="balanced", random_state=42)),
    ]),
    "RF + ROS": ImbPipeline([
        ("preprocess", preprocessor),
        ("oversample", RandomOverSampler(random_state=42)),
        ("model", RandomForestClassifier(n_estimators=40, max_depth=4, min_samples_leaf=3, random_state=42)),
    ]),
}

result_oversampling = []
for model_name, pipe in oversampling_models.items():
    scores = cross_validate(pipe, X, y_binary, scoring=scoring, cv=cv, n_jobs=1)
    result_oversampling.append({
        "experiment": "Binary At Risk with Oversampling",
        "model": model_name,
        "accuracy_mean": scores["test_accuracy"].mean(),
        "balanced_accuracy_mean": scores["test_balanced_accuracy"].mean(),
        "f1_macro_mean": scores["test_f1_macro"].mean(),
        "f1_weighted_mean": scores["test_f1_weighted"].mean(),
    })

result_oversampling = pd.DataFrame(result_oversampling).sort_values("f1_macro_mean", ascending=False)
result_oversampling.round(3)
```

## 9. Leakage Demo — Skor Tinggi tapi Tidak Valid

Bagian ini sengaja memasukkan `depression`, `anxiety`, dan `panic_attack` sebagai fitur untuk memprediksi `risk_level`. Karena `risk_level` dibuat dari tiga indikator itu, hasilnya sangat tinggi, tetapi tidak valid untuk diklaim sebagai model prediksi.

```python
LEAK_FEATURES = FEATURES + ["depression", "anxiety", "panic_attack"]
X_leak = df[LEAK_FEATURES]

leak_preprocessor = ColumnTransformer([
    ("num", Pipeline([
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler())
    ]), NUM_FEATURES),
    ("cat", Pipeline([
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
    ]), CAT_FEATURES + ["depression", "anxiety", "panic_attack"]),
])

leak_models = {
    "LogReg Leakage": LogisticRegression(max_iter=1000, random_state=42),
    "RF Leakage": RandomForestClassifier(n_estimators=40, max_depth=4, min_samples_leaf=3, random_state=42),
}

leak_rows = []
for model_name, model in leak_models.items():
    pipe = Pipeline([
        ("preprocess", leak_preprocessor),
        ("model", model)
    ])
    scores = cross_validate(pipe, X_leak, y_multi, scoring=scoring, cv=cv, n_jobs=1)
    leak_rows.append({
        "experiment": "Leakage Demo - Invalid",
        "model": model_name,
        "accuracy_mean": scores["test_accuracy"].mean(),
        "balanced_accuracy_mean": scores["test_balanced_accuracy"].mean(),
        "f1_macro_mean": scores["test_f1_macro"].mean(),
        "f1_weighted_mean": scores["test_f1_weighted"].mean(),
    })

result_leakage = pd.DataFrame(leak_rows).sort_values("f1_macro_mean", ascending=False)
result_leakage.round(3)
```

## 10. Holdout Report Model Terbaik

Holdout split tetap ditampilkan sebagai gambaran tambahan, tetapi kesimpulan utama tetap memakai cross-validation karena dataset kecil.

```python
def holdout_report(X, y, model, title):
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.25, random_state=42, stratify=y
    )
    pipe = Pipeline([
        ("preprocess", preprocessor),
        ("model", model)
    ])
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)

    print(title)
    print("Accuracy:", round(accuracy_score(y_test, y_pred), 3))
    print("Balanced Accuracy:", round(balanced_accuracy_score(y_test, y_pred), 3))
    print("F1 Macro:", round(f1_score(y_test, y_pred, average="macro"), 3))
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred, digits=3))

best_multi_model_name = result_multi.iloc[0]["model"]
best_binary_model_name = result_binary.iloc[0]["model"]

print("Best multiclass model from CV:", best_multi_model_name)
print("Best binary model from CV:", best_binary_model_name)
print("-" * 70)
holdout_report(X, y_multi, models[best_multi_model_name], f"Multiclass Holdout — {best_multi_model_name}")
print("-" * 70)
holdout_report(X, y_binary, models[best_binary_model_name], f"Binary Holdout — {best_binary_model_name}")
```

## 11. Ringkasan Akhir Eksperimen Valid

```python
all_valid_results = pd.concat([result_multi, result_binary, result_oversampling], ignore_index=True)
all_valid_results_sorted = all_valid_results.sort_values("f1_macro_mean", ascending=False)
all_valid_results_sorted.round(3)
```

## 12. Feature Importance — Model Valid

Bagian ini tetap memakai pipeline dan daftar model dari eksperimen sebelumnya. Tujuannya adalah melihat fitur mana yang paling sering dianggap penting oleh model valid.

Catatan penting: fitur `depression`, `anxiety`, dan `panic_attack` **tidak dimasukkan** pada bagian ini, karena ketiganya dipakai untuk membentuk target `risk_level` dan `at_risk`. Kalau dimasukkan, model akan mengalami data leakage.

```python
from sklearn.base import clone
from sklearn.inspection import permutation_importance


def get_feature_names_from_pipeline(fitted_pipe, feature_cols):
    """Mengambil nama fitur setelah preprocessing OneHotEncoder."""
    pre = fitted_pipe.named_steps["preprocess"]
    feature_names = []

    # Numeric features
    if "num" in pre.named_transformers_:
        feature_names.extend(NUM_FEATURES)

    # Categorical features after one-hot encoding
    if "cat" in pre.named_transformers_:
        cat_pipeline = pre.named_transformers_["cat"]
        encoder = cat_pipeline.named_steps["onehot"]
        cat_names = encoder.get_feature_names_out(CAT_FEATURES)
        feature_names.extend(cat_names)

    return list(feature_names)


def map_base_feature(encoded_feature_name, base_features):
    """Mengembalikan nama fitur asli dari fitur hasil one-hot encoding."""
    for base in sorted(base_features, key=len, reverse=True):
        if encoded_feature_name == base or encoded_feature_name.startswith(base + "_"):
            return base
    return encoded_feature_name


def extract_builtin_importance(fitted_pipe, base_features, experiment_name, model_name):
    """Mengambil feature importance/coef dari model yang mendukung."""
    model = fitted_pipe.named_steps["model"]
    feature_names = get_feature_names_from_pipeline(fitted_pipe, base_features)

    if hasattr(model, "feature_importances_"):
        values = np.asarray(model.feature_importances_)
        importance_type = "tree_feature_importance"
    elif hasattr(model, "coef_"):
        coef = np.asarray(model.coef_)
        if coef.ndim == 1:
            values = np.abs(coef)
        else:
            values = np.mean(np.abs(coef), axis=0)
        importance_type = "absolute_coefficient"
    else:
        return pd.DataFrame()

    n = min(len(feature_names), len(values))
    temp = pd.DataFrame({
        "experiment": experiment_name,
        "model": model_name,
        "encoded_feature": feature_names[:n],
        "importance": values[:n],
        "importance_type": importance_type,
    })
    temp["base_feature"] = temp["encoded_feature"].apply(lambda x: map_base_feature(x, base_features))

    # Normalisasi supaya mudah dibandingkan antar model
    total = temp["importance"].sum()
    if total > 0:
        temp["importance_norm"] = temp["importance"] / total
    else:
        temp["importance_norm"] = 0

    return temp


importance_frames = []

# 1) Feature importance untuk semua model valid non-oversampling
for experiment_name, target in [
    ("Multiclass Risk Level", y_multi),
    ("Binary At Risk", y_binary),
]:
    for model_name, model in models.items():
        if model_name == "Dummy Most Frequent":
            continue

        pipe = Pipeline([
            ("preprocess", preprocessor),
            ("model", clone(model)),
        ])
        pipe.fit(X, target)
        fi = extract_builtin_importance(pipe, FEATURES, experiment_name, model_name)
        if not fi.empty:
            importance_frames.append(fi)

# 2) Feature importance untuk model oversampling binary
for model_name, pipe in oversampling_models.items():
    fitted_pipe = clone(pipe)
    fitted_pipe.fit(X, y_binary)
    fi = extract_builtin_importance(
        fitted_pipe,
        FEATURES,
        "Binary At Risk with Oversampling",
        model_name,
    )
    if not fi.empty:
        importance_frames.append(fi)

valid_feature_importance_detail = pd.concat(importance_frames, ignore_index=True)

valid_feature_importance_summary = (
    valid_feature_importance_detail
    .groupby(["experiment", "model", "base_feature", "importance_type"], as_index=False)
    .agg(importance_norm=("importance_norm", "sum"))
    .sort_values(["experiment", "model", "importance_norm"], ascending=[True, True, False])
)

valid_feature_importance_summary.head(30).round(4)
```

### 12.1 Rata-rata Feature Importance dari Semua Model Valid

Tabel di bawah merangkum fitur mana yang secara konsisten muncul sebagai fitur penting dari berbagai model valid. Nilainya bukan untuk klaim klinis, tetapi untuk membaca pola teknis dari model.

```python
overall_valid_importance = (
    valid_feature_importance_summary
    .groupby("base_feature", as_index=False)
    .agg(
        mean_importance=("importance_norm", "mean"),
        max_importance=("importance_norm", "max"),
        appeared_in_models=("model", "nunique"),
    )
    .sort_values("mean_importance", ascending=False)
)

overall_valid_importance.round(4)
```

```python
import plotly.express as px

fig = px.bar(
    overall_valid_importance,
    x="mean_importance",
    y="base_feature",
    orientation="h",
    title="Average Feature Importance dari Semua Model Valid",
    labels={"mean_importance": "Mean Normalized Importance", "base_feature": "Base Feature"},
)
fig.update_layout(height=430, yaxis={"categoryorder": "total ascending"})
fig.show()
```

### 12.2 Permutation Importance untuk Model Binary Terbaik

Permutation importance dipakai sebagai pengecekan tambahan karena lebih model-agnostic. Prinsipnya: satu fitur diacak, lalu dilihat seberapa besar performa model turun. Jika performa turun besar, fitur tersebut dianggap lebih penting.

Bagian ini memakai model binary terbaik dari hasil cross-validation karena target binary lebih cocok untuk skenario early warning awal.

```python
# Ambil model binary valid terbaik dari hasil eksperimen sebelumnya
binary_valid_candidates = all_valid_results_sorted[
    all_valid_results_sorted["experiment"].str.contains("Binary", case=False, regex=False)
].copy()

best_binary_row = binary_valid_candidates.iloc[0]
best_binary_experiment = best_binary_row["experiment"]
best_binary_model_name = best_binary_row["model"]

print("Best binary experiment:", best_binary_experiment)
print("Best binary model:", best_binary_model_name)

# Build pipeline sesuai asal eksperimennya
if best_binary_experiment == "Binary At Risk with Oversampling":
    best_binary_pipe = clone(oversampling_models[best_binary_model_name])
else:
    best_binary_pipe = Pipeline([
        ("preprocess", preprocessor),
        ("model", clone(models[best_binary_model_name])),
    ])

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y_binary,
    test_size=0.25,
    random_state=42,
    stratify=y_binary,
)

best_binary_pipe.fit(X_train, y_train)
y_pred = best_binary_pipe.predict(X_test)

print("Holdout Accuracy:", round(accuracy_score(y_test, y_pred), 3))
print("Holdout Balanced Accuracy:", round(balanced_accuracy_score(y_test, y_pred), 3))
print("Holdout F1 Macro:", round(f1_score(y_test, y_pred, average="macro"), 3))

perm = permutation_importance(
    best_binary_pipe,
    X_test,
    y_test,
    scoring="f1_macro",
    n_repeats=30,
    random_state=42,
)

permutation_importance_df = pd.DataFrame({
    "feature": FEATURES,
    "importance_mean": perm.importances_mean,
    "importance_std": perm.importances_std,
}).sort_values("importance_mean", ascending=False)

permutation_importance_df.round(4)
```

```python
fig = px.bar(
    permutation_importance_df,
    x="importance_mean",
    y="feature",
    error_x="importance_std",
    orientation="h",
    title=f"Permutation Importance — {best_binary_model_name}",
    labels={"importance_mean": "Mean Decrease in F1 Macro", "feature": "Feature"},
)
fig.update_layout(height=430, yaxis={"categoryorder": "total ascending"})
fig.show()
```

## 13. Feature Importance — Leakage Demo

Bagian ini sengaja memasukkan `depression`, `anxiety`, dan `panic_attack` sebagai fitur. Tujuannya bukan untuk membuat klaim model, tetapi untuk membuktikan kenapa skor bisa menjadi sangat tinggi ketika fitur pembentuk target ikut dimasukkan.

Jika tiga fitur ini muncul sebagai fitur paling dominan, berarti model sebenarnya hanya membaca jawaban yang sudah membentuk target.

```python
def get_leak_feature_names_from_pipeline(fitted_pipe):
    pre = fitted_pipe.named_steps["preprocess"]
    feature_names = []

    if "num" in pre.named_transformers_:
        feature_names.extend(NUM_FEATURES)

    if "cat" in pre.named_transformers_:
        leak_cat_features = CAT_FEATURES + ["depression", "anxiety", "panic_attack"]
        cat_pipeline = pre.named_transformers_["cat"]
        encoder = cat_pipeline.named_steps["onehot"]
        cat_names = encoder.get_feature_names_out(leak_cat_features)
        feature_names.extend(cat_names)

    return list(feature_names)


def extract_leak_importance(fitted_pipe, base_features, experiment_name, model_name):
    model = fitted_pipe.named_steps["model"]
    feature_names = get_leak_feature_names_from_pipeline(fitted_pipe)

    if hasattr(model, "feature_importances_"):
        values = np.asarray(model.feature_importances_)
        importance_type = "tree_feature_importance"
    elif hasattr(model, "coef_"):
        coef = np.asarray(model.coef_)
        if coef.ndim == 1:
            values = np.abs(coef)
        else:
            values = np.mean(np.abs(coef), axis=0)
        importance_type = "absolute_coefficient"
    else:
        return pd.DataFrame()

    n = min(len(feature_names), len(values))
    temp = pd.DataFrame({
        "experiment": experiment_name,
        "model": model_name,
        "encoded_feature": feature_names[:n],
        "importance": values[:n],
        "importance_type": importance_type,
    })
    temp["base_feature"] = temp["encoded_feature"].apply(lambda x: map_base_feature(x, base_features))

    total = temp["importance"].sum()
    if total > 0:
        temp["importance_norm"] = temp["importance"] / total
    else:
        temp["importance_norm"] = 0

    return temp


leak_importance_frames = []
leak_base_features = LEAK_FEATURES

for model_name, model in leak_models.items():
    pipe = Pipeline([
        ("preprocess", leak_preprocessor),
        ("model", clone(model)),
    ])
    pipe.fit(X_leak, y_multi)
    fi = extract_leak_importance(pipe, leak_base_features, "Leakage Demo - Invalid", model_name)
    if not fi.empty:
        leak_importance_frames.append(fi)

leak_feature_importance_detail = pd.concat(leak_importance_frames, ignore_index=True)

leak_feature_importance_summary = (
    leak_feature_importance_detail
    .groupby(["model", "base_feature", "importance_type"], as_index=False)
    .agg(importance_norm=("importance_norm", "sum"))
    .sort_values(["model", "importance_norm"], ascending=[True, False])
)

leak_feature_importance_summary.round(4)
```

```python
overall_leak_importance = (
    leak_feature_importance_summary
    .groupby("base_feature", as_index=False)
    .agg(mean_importance=("importance_norm", "mean"))
    .sort_values("mean_importance", ascending=False)
)

overall_leak_importance.round(4)
```

```python
fig = px.bar(
    overall_leak_importance,
    x="mean_importance",
    y="base_feature",
    orientation="h",
    title="Average Feature Importance pada Leakage Demo — Invalid",
    labels={"mean_importance": "Mean Normalized Importance", "base_feature": "Base Feature"},
)
fig.update_layout(height=430, yaxis={"categoryorder": "total ascending"})
fig.show()
```

## 14. Interpretasi Feature Importance

Dari feature importance model valid, fitur seperti `course`, `cgpa`, `year_of_study`, `age`, `gender`, dan `marital_status` hanya memberi sinyal tidak langsung. Karena itu, skor model valid tidak bisa melonjak terlalu tinggi.

Sebaliknya, pada leakage demo, fitur `depression`, `anxiety`, dan `panic_attack` muncul sangat dominan karena target `risk_level` memang dibuat dari tiga indikator tersebut. Ini membuktikan bahwa skor tinggi pada leakage bukan performa prediksi yang valid, melainkan efek kebocoran data.

Implikasi untuk solusi MindTrack UB:

1. Model ML tetap diposisikan sebagai **proof of concept**.
2. Risk scoring berbasis indikator mental health lebih tepat dipakai sebagai mekanisme segmentasi awal.
